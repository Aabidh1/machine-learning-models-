```python
plt.figure(figsize=(10,5))
sns.countplot(x='sales', data=df, order=df['sales'].value_counts().index)
plt.xticks(rotation=45)
plt.title("Distribution of Employees by Department")
plt.show()
```

```python
df['salary'].value_counts().plot.pie(autopct="%1.1f%%", colors=["red", "yellow", "green"])
plt.title("Salary Level Distribution")
plt.ylabel("")
plt.show()
```

```python
num_cols = ["satisfaction_level", "last_evaluation", "number_project", 
            "average_montly_hours", "time_spend_company"]

plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)  
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="salary", y="satisfaction_level", data=df, palette="coolwarm")
plt.title("Satisfaction Level Across Salary Categories")
plt.show()
```

```python
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)  # Subplot grid
    sns.histplot(df[col], kde=True, bins=30, color="blue")  
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python

num_cols = ["satisfaction_level", "last_evaluation", "number_project", 
            "average_montly_hours", "time_spend_company"]


corr_matrix = df[num_cols].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Variables")
plt.show()
```

```python
sns.countplot(x="salary", hue="left", data=df, palette="Set1")
plt.title("Attrition Rate by Salary")
plt.show()
```

```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(df["sales"], df["left"])
chi2, p, _, _ = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2}, p-value: {p}")
```
